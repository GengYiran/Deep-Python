```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('classic')
x = np.linspace(0, 10, 100)  # from 0 to 10 get 100 pieces
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y, '-')
plt.plot(x, z, '--')
plt.xlabel('x_lable')
plt.ylabel('y_label')
plt.title('title')
plt.legend(['sin', 'cos'])
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)[:, np.newaxis]
z = np.sin(2*x)+np.cos(2*y)
print(x.shape, y.shape, z.shape)
plt.imshow(z)

```

```python
x = np.arange(100)
y = np.multiply.outer(x, x)
print(y.shape)
plt.imshow(y)
```
